```python
import pandas as pd
import numpy as np
import datetime
```

```python
jan_2018 = "data/2018/jan_2018.csv"
feb_2018 = "data/2018/feb_2018.csv"
mar_2018 = "data/2018/mar_2018.csv"
apr_2018 = "data/2018/apr_2018.csv"
may_2018 = "data/2018/may_2018.csv"
june_2018 = "data/2018/june_2018.csv"
july_2018 = "data/2018/july_2018.csv"
aug_2018 = "data/2018/aug_2018.csv"
sept_2018 = "data/2018/sept_2018.csv"
oct_2018 = "data/2018/oct_2018.csv"
nov_2018 = "data/2018/nov_2018.csv"
dec_2018 = "data/2018/dec_2018.csv"

```

```python
jan_2018_df = pd.read_csv(jan_2018)
feb_2018_df = pd.read_csv(feb_2018)
mar_2018_df = pd.read_csv(mar_2018)
apr_2018_df = pd.read_csv(apr_2018)
may_2018_df = pd.read_csv(may_2018)
june_2018_df = pd.read_csv(june_2018)
july_2018_df = pd.read_csv(july_2018)
aug_2018_df = pd.read_csv(aug_2018)
sept_2018_df = pd.read_csv(sept_2018)
oct_2018_df = pd.read_csv(oct_2018)
nov_2018_df = pd.read_csv(nov_2018)
dec_2018_df = pd.read_csv(dec_2018)

jan_2018_df.head()
```

```python
citibike_2018_df = pd.concat([jan_2018_df, feb_2018_df, mar_2018_df, apr_2018_df, may_2018_df, june_2018_df, july_2018_df, aug_2018_df, sept_2018_df, oct_2018_df, nov_2018_df, dec_2018_df ])

citibike_2018_df.head()
```

```python
citibike_2018_df['Dates'] = pd.to_datetime(citibike_2018_df['starttime']).dt.date
citibike_2018_df['startTime'] = pd.to_datetime(citibike_2018_df['starttime']).dt.time
citibike_2018_df['endTime'] = pd.to_datetime(citibike_2018_df['stoptime']).dt.time
citibike_2018_df.head()
```

```python
citibike_2018_df['Year'] = pd.to_datetime(citibike_2018_df['Dates']).dt.year
citibike_2018_df.head()
```

```python
citibike_2018_df['Day of Week'] = pd.to_datetime(citibike_2018_df['Dates']).dt.weekday_name
citibike_2018_df.head()
```

```python
citibike_2018_df.loc[(citibike_2018_df['gender'] == 1, 'gender')] = 'Male'
citibike_2018_df.loc[(citibike_2018_df['gender'] == 2, 'gender')] = 'Female'
citibike_2018_df.loc[(citibike_2018_df['gender'] == 0, 'gender')] = 'Unknown'
citibike_2018_df.head()
```

```python
citibike_2018_df['tripDuration_minutes'] = citibike_2018_df['tripduration']/60
citibike_2018_df['tripDuration_minutes'] = citibike_2018_df['tripDuration_minutes'].round(decimals = 2)
citibike_2018_df.head()
```

```python
citibike_2018_df.to_csv("data/2018/citibike_2018.csv")
```
