```python
import pandas as pd

citibike_2017 = "data/2017/citibike_2017.csv"
citibike_2018 = "data/2018/citibike_2018.csv"
citibike_2019 = "data/2019/citibike_2019.csv"

citibike_2017_df = pd.read_csv(citibike_2017)
citibike_2018_df = pd.read_csv(citibike_2018)
citibike_2019_df = pd.read_csv(citibike_2019)

citibike_2019_df.head()
```

```python
citibike_2017thru2019_df = pd.concat([citibike_2017_df, citibike_2018_df, citibike_2019_df])

citibike_2017thru2019_df.head()
```

```python
citibike_2017thru2019_df.loc[:,"tripDuration_minutes"].median()
```

```python
citibike_2017thru2019_df["tripDuration_minutes"].mean()
```

```python

citibike_2017thru2019_df["tripDuration_minutes"].describe()
```

```python
citibike_2017thru2019_df.tripDuration_minutes = citibike_2017thru2019_df.tripDuration_minutes.round()
citibike_2017thru2019_df.head()
```

```python
citibike_2017thru2019_df.to_csv("data/citibike_2017-2019.csv")
```
