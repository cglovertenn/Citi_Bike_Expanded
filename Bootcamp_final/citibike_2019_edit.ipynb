```python
import pandas as pd
import numpy as np
import datetime
```

```python
jan_2019 = "data/2019/jan_2019.csv"
feb_2019 = "data/2019/feb_2019.csv"
mar_2019 = "data/2019/mar_2019.csv"
apr_2019 = "data/2019/apr_2019.csv"
may_2019 = "data/2019/may_2019.csv"
june_2019 = "data/2019/june_2019.csv"
july_2019 = "data/2019/july_2019.csv"
aug_2019 = "data/2019/aug_2019.csv"
sept_2019 = "data/2019/sept_2019.csv"
oct_2019 = "data/2019/oct_2019.csv"
nov_2019 = "data/2019/nov_2019.csv"
dec_2019 = "data/2019/dec_2019.csv"
```

```python
jan_2019_df = pd.read_csv(jan_2019)
feb_2019_df = pd.read_csv(feb_2019)
mar_2019_df = pd.read_csv(mar_2019)
apr_2019_df = pd.read_csv(apr_2019)
may_2019_df = pd.read_csv(may_2019)
june_2019_df = pd.read_csv(june_2019)
july_2019_df = pd.read_csv(july_2019)
aug_2019_df = pd.read_csv(aug_2019)
sept_2019_df = pd.read_csv(sept_2019)
oct_2019_df = pd.read_csv(oct_2019)
nov_2019_df = pd.read_csv(nov_2019)
dec_2019_df = pd.read_csv(dec_2019)

jan_2019_df.head()
```

```python
citibike_2019_df = pd.concat([jan_2019_df, feb_2019_df, mar_2019_df, apr_2019_df, may_2019_df, june_2019_df, july_2019_df, aug_2019_df, sept_2019_df, oct_2019_df, nov_2019_df, dec_2019_df ])

citibike_2019_df.head()
```

```python
citibike_2019_df['Dates'] = pd.to_datetime(citibike_2019_df['starttime']).dt.date
citibike_2019_df['startTime'] = pd.to_datetime(citibike_2019_df['starttime']).dt.time
citibike_2019_df['endTime'] = pd.to_datetime(citibike_2019_df['stoptime']).dt.time
citibike_2019_df.head()
```

```python
citibike_2019_df['Year'] = pd.to_datetime(citibike_2019_df['Dates']).dt.year
citibike_2019_df.head()
```

```python
citibike_2019_df['Day of Week'] = pd.to_datetime(citibike_2019_df['Dates']).dt.weekday_name
citibike_2019_df.head()
```

```python
citibike_2019_df.loc[(citibike_2019_df['gender'] == 1, 'gender')] = 'Male'
citibike_2019_df.loc[(citibike_2019_df['gender'] == 2, 'gender')] = 'Female'
citibike_2019_df.loc[(citibike_2019_df['gender'] == 0, 'gender')] = 'Unknown'
citibike_2019_df.head()
```

```python
citibike_2019_df['tripDuration_minutes'] = citibike_2019_df['tripduration']/60
citibike_2019_df['tripDuration_minutes'] = citibike_2019_df['tripDuration_minutes'].round(decimals = 2)
citibike_2019_df.head()
```

```python
citibike_2019_df.to_csv("data/2019/citibike_2019.csv")
```
