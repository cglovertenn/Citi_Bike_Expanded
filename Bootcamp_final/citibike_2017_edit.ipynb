```python
import pandas as pd
import numpy as np
import datetime
```

```python
jan_2017 = "data/2017/jan_2017.csv"
feb_2017 = "data/2017/feb_2017.csv"
mar_2017 = "data/2017/mar_2017.csv"
apr_2017 = "data/2017/apr_2017.csv"
may_2017 = "data/2017/may_2017.csv"
june_2017 = "data/2017/june_2017.csv"
july_2017 = "data/2017/july_2017.csv"
aug_2017 = "data/2017/aug_2017.csv"
sept_2017 = "data/2017/sept_2017.csv"
oct_2017 = "data/2017/oct_2017.csv"
nov_2017 = "data/2017/nov_2017.csv"
dec_2017 = "data/2017/dec_2017.csv"
```

```python
jan_2017_df = pd.read_csv(jan_2017)
feb_2017_df = pd.read_csv(feb_2017)
mar_2017_df = pd.read_csv(mar_2017)
apr_2017_df = pd.read_csv(apr_2017)
may_2017_df = pd.read_csv(may_2017)
june_2017_df = pd.read_csv(june_2017)
july_2017_df = pd.read_csv(july_2017)
aug_2017_df = pd.read_csv(aug_2017)
sept_2017_df = pd.read_csv(sept_2017)
oct_2017_df = pd.read_csv(oct_2017)
nov_2017_df = pd.read_csv(nov_2017)
dec_2017_df = pd.read_csv(dec_2017)

dec_2017_df.head()
```

```python
citibike_2017_df = pd.concat([jan_2017_df, feb_2017_df, mar_2017_df, apr_2017_df, may_2017_df, june_2017_df, july_2017_df,
                             aug_2017_df, sept_2017_df, oct_2017_df, nov_2017_df, dec_2017_df])
citibike_2017_df.head()
```

```python
citibike_2017_df['Dates'] = pd.to_datetime(citibike_2017_df['starttime']).dt.date
citibike_2017_df['startTime'] = pd.to_datetime(citibike_2017_df['starttime']).dt.time
citibike_2017_df['endTime'] = pd.to_datetime(citibike_2017_df['stoptime']).dt.time
citibike_2017_df.head()
```

```python
citibike_2017_df['Year'] = pd.to_datetime(citibike_2017_df['Dates']).dt.year
citibike_2017_df.head()
```

```python
citibike_2017_df['Day of Week'] = pd.to_datetime(citibike_2017_df['Dates']).dt.weekday_name
citibike_2017_df.head()
```

```python
citibike_2017_df.loc[(citibike_2017_df['gender'] == 1, 'gender')] = 'Male'
citibike_2017_df.loc[(citibike_2017_df['gender'] == 2, 'gender')] = 'Female'
citibike_2017_df.loc[(citibike_2017_df['gender'] == 0, 'gender')] = 'Unknown'
citibike_2017_df.head()
```

```python
citibike_2017_df['tripDuration_minutes'] = citibike_2017_df['tripduration']/60
citibike_2017_df['tripDuration_minutes'] = citibike_2017_df['tripDuration_minutes'].round(decimals = 2)
citibike_2017_df.head()
```

```python
citibike_2017_df.to_csv("data/2017/citibike_2017.csv")
```
